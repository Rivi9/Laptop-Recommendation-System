```python
import spacy as sc
```

loading the spacy english trained model

```python
nlp = sc.load("en_core_web_sm")
```

Preprocess the 'description' column using tokenizing method

```python
def preprocess(text):
  doc = nlp(text)
  filtered_tokens = []
  for token in doc:
    if token.is_stop or token.is_punct:
      continue
    else:
      filtered_tokens.append(token.lemma_)
  return " ".join(filtered_tokens)
```

```python
df['preprocess-text'] = df['Para'].apply(preprocess)
```

## Splitting data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df['Para'], df['name'], test_size=0.2, random_state=42)
```

Logistice Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer()
X_train_cv = cv.fit_transform(x_train)
X_test_cv = cv.transform(x_test)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=100)
rf.fit(X_train_cv, y_train)
```

```python
y_pred_rf = rf.predict(X_test_cv)
```

```python
from sklearn.metrics import classification_report
```

```python
report_rf = classification_report(y_test, y_pred_rf)
print("Classification report of Random Forest classifier:\n", report_rf)
```
