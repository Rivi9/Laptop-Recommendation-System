```python
laptops['processor'] = laptops['processor'].apply(lambda x:x.split())
laptops['ram'] = laptops['ram'].apply(lambda x:x.split())
laptops['os'] = laptops['os'].apply(lambda x:x.split())
laptops['use'] = laptops['usecases'].apply(lambda x:x.split())
```

```python
laptops['tags'] = laptops['processor']+laptops['ram']+laptops['os']+laptops['use']
```

```python
new = laptops.drop(columns=['processor','ram','os','storage','rating','os_brand','processor_brand','use'])
```

```python
new.head(2)
```

```python
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
```

```python
new.iloc[0].tags
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head(2)
```

```python
new.iloc[0].tags
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new.head(2)
```

```python
import nltk
nltk.download('punkt')

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Tokenize the tags
new['tags_tokens'] = new['tags'].apply(word_tokenize)

# Train the Word2Vec model
model = Word2Vec(sentences=new['tags_tokens'], vector_size=100, window=5, min_count=1, workers=4)

# Get the Word2Vec vectors for each laptop
word2vec_vectors = []
for tags_tokens in new['tags_tokens']:
    vectors = [model.wv[word] for word in tags_tokens if word in model.wv]
    if vectors:
        word2vec_vectors.append(np.mean(vectors, axis=0))
    else:
        word2vec_vectors.append(np.zeros(100))
word2vec_vectors = np.array(word2vec_vectors)
```

```python
# counter vectorizing
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
#Recommend
```

```python
def recommend(laptop):
    index = new[new['usecases']==laptop].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x:x[1])
    for i in distances[1:10]:
        print(new.iloc[i[0]].usecases)
```

```python
def recommend(use):
    index = new[new['usecases'] == use].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    printed_names = set()
    for i in distances[1:50]:
        row_index = i[0]
        name = new.iloc[row_index]['name']
        if name not in printed_names:
            printed_names.add(name)
            print(name)
```
